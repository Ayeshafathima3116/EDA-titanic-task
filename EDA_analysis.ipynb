```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Titanic-Dataset.csv")  # or your cleaned version
df.head()
```

```python
df.info()
df.describe()
```

```python
df['Age'].hist(bins=30, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

df['Fare'].hist(bins=30, edgecolor='black')
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age")
plt.show()

sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.show()
```

 Observations from EDA

- **Younger passengers** generally had a higher survival rate.
- **Female passengers** had a significantly higher survival rate compared to males.
- Passengers in **1st class** had better survival chances than those in 2nd or 3rd class.
- The **Fare** column has some outliers with very high ticket prices.
- The **Age** distribution is slightly **right-skewed**, meaning more passengers were younger.
- There is a moderate correlation between **Fare** and **Pclass**, and some features are weakly correlated.
